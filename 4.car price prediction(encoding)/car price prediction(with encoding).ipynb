```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car.csv')
df
```

### Now here we creating dummies of car models for one hot encoding

```python
dummy = pd.get_dummies(df['Car Model'])
dummy
```

### now We concate two dataframe 

```python
concat_df = pd.concat([df, dummy], axis='columns')
concat_df
```

```python
final_df = concat_df.drop(['Car Model', 'Mercedez'], axis='columns')
final_df
```

### now create model

```python
'''
from sklearn import linear_model
reg = linear_model.LinearRegression()

'''

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = final_df.drop(['Sell Price($)'], axis = 'columns')
```

```python
X
```

```python
y = final_df['Sell Price($)']
```

```python
y
```

```python
model.fit(X, y)
```

###### Price of mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,0,0]])
```

##### Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X,y)*100
```
